```python
import pandas as pd
import numpy as np

# 读取文件
df = pd.read_csv('fig5_eval_1772341471.csv')

# --- 计算 Heading 误差 ---
# 论文定义：Heading Φ = |arctan2(uy, ux)|
# ux, uy 是你在脚本中记录的指向目标的意图方向
df['heading_error'] = np.arctan2(df['uy'], df['ux'])
# 如果您记录的是意图向量与实际航向的夹角，计算公式可能需要调整
# 根据您之前提供的代码逻辑，您记录的是相对于机器人的意图方向。

# --- 计算 Jerk (平滑度) ---
# Jerk 是加速度的导数。首先计算加速度，再计算 Jerk。
dt = df['timestamp'].diff()
df['accel_v'] = df['vel_v'].diff() / dt
df['accel_w'] = df['vel_w'].diff() / dt

df['jerk_v'] = df['accel_v'].diff() / dt
df['jerk_w'] = df['accel_w'].diff() / dt

# 计算总 Jerk 幅值
df['total_jerk'] = np.sqrt(df['jerk_v']**2 + df['jerk_w']**2)
```
